# Data Wrangling:

Replacing missing values from dataset features:

```python
features["CPI"].fillna(features["CPI"].median(),inplace=True)
features["Unemployment"].fillna(features["Unemployment"].median(),inplace=True)
```

```python
features.isnull().sum()
```

```python
for i in range(1,6):
  features["MarkDown"+str(i)] = features["MarkDown"+str(i)].apply(lambda x: 0 if x < 0 else x)
  features["MarkDown"+str(i)].fillna(value=0,inplace=True)
```

```python
features.isnull().sum()
```

```python
data = pd.merge(data,stores,on='Store',how='left')
```

```python
data = pd.merge(data,features,on=['Store','Date'],how='left')
```

```python
data.sort_values(by=['Date'],inplace=True)
```

```python
data.set_index(data.Date, inplace=True)
```

```python
data['IsHoliday_x'].isin(data['IsHoliday_y']).all()
```

```python
data.drop(columns='IsHoliday_x',inplace=True)
data.rename(columns={"IsHoliday_y" : "IsHoliday"}, inplace=True)
data.info()
```

```python
data['Date'] = pd.to_datetime(data['Date'], format='%d-%m-%Y')
```

```python
data.head(5)
```

```python
data.info()
```

```python
data['Year'] = data['Date'].dt.year
data['Month'] = data['Date'].dt.month
data['Day'] = data['Date'].dt.day
```

```python
data.head(5)
```

```python
agg_data = data.groupby(['Store', 'Dept']).Weekly_Sales.agg(['max', 'min', 'mean', 'median', 'std']).reset_index()
agg_data.isnull().sum()
```

```python
store_data = pd.merge(left=data,right=agg_data,on=['Store', 'Dept'],how ='left')
store_data.dropna(inplace=True)
data = store_data.copy()
del store_data
```

```python
data['Date'] = pd.to_datetime(data['Date'])
data.sort_values(by=['Date'],inplace=True)
data.set_index(data.Date, inplace=True)
data.head()
```

```python
data['Total_MarkDown'] = data['MarkDown1']+data['MarkDown2']+data['MarkDown3']+data['MarkDown4']+data['MarkDown5']
data.drop(['MarkDown1','MarkDown2','MarkDown3','MarkDown4','MarkDown5'], axis = 1,inplace=True)
```

```python
data=data[data['Weekly_Sales']>=0]
```

```python
data.shape
```

```python
data['IsHoliday']=data['IsHoliday'].astype('int')
```

```python
data
```

```python
data.to_csv('./clean.walmart_dataset.csv')
```

# Data Transformation:

Getting dummy values:

```python
dummies_col = ['Store','Dept','Type']
data_dummies = data[dummies_col].copy()
```

```python
data_dummies.head(10)
```

```python
data_dummies = pd.get_dummies(data_dummies).astype('int')
```

```python
data_dummies.head(5)
```

```python
data = pd.concat([data, data_dummies],axis=1)
```

```python
data.drop(columns=dummies_col,inplace=True)
data.drop(columns=['Date'],inplace=True)
```

# Traning and Testing the Model:

Spliting the datasets:

```python
X = data.drop(['Weekly_Sales'],axis=1)
Y = data.Weekly_Sales
X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size=0.20, random_state=50)
```

Using Linear Regression Model:

```python
model_lr = LinearRegression()
model_lr.fit(X_train, y_train)
```

Calculating Accuracy:

```python
model_lr_acc = model_lr.score(X_test,y_test)*100
print("Linear Regressor Accuracy is: ",model_lr_acc)
```

Plotting graph of predictions:

```python
plt.figure(figsize=(18,12))
plt.plot(model_lr.predict(X_test[:100]), label="prediction", linewidth=2.0,color='blue')
plt.plot(y_test[:100].values, label="real_values", linewidth=2.0,color='cyan')
plt.legend(loc="best")
plt.savefig('graphs/modellr1_real_pred.png')
plt.show()
```
